# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF -using the equation- of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

```python

# Step 1: Generate a standard Gaussian random variable (zero mean, unit variance)  
X = np.random.normal(loc=0, scale=1, size=10000)

# Step 2: Plot the histogram (estimated PDF) of X
plt.figure(figsize=(10, 6))
count, bins, _ = plt.hist(X, bins=50, density=True, alpha=0.6,  label="Estimated PDF (Histogram)")

# Step 3: Compute and plot the true analytical PDF of X
x_vals = np.linspace(-4, 4, 1000)
true_pdf = norm.pdf(x_vals, loc=0, scale=1)
plt.plot(x_vals, true_pdf, color='red', lw=2, label="True Analytical PDF")

# Step 4: Calculate the sample mean and sample variance
sample_mean = np.mean(X)
sample_variance = np.var(X)

# Display results in the required format
plt.title("Histogram and Analytical PDF of Standard Gaussian Random Variable")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)

# Print sample mean and variance
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Display the plot
plt.show()
```

- Compute the CDF using [scipy.stats.norm.cdf](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), then plot it.

```python
# # Step 1: Generate a standard Gaussian random variable (zero mean, unit variance)
# X = np.random.normal(loc=0, scale=1, size=10000)

# Step 2: Plot the histogram (estimated PDF) of X
plt.figure(figsize=(12, 8))
count, bins, _ = plt.hist(X, bins=50, density=True, alpha=0.6, label="Estimated PDF (Histogram)")

# Step 3: Compute and plot the true analytical PDF of X
x_vals = np.linspace(-4, 4, 1000)
true_pdf = norm.pdf(x_vals, loc=0, scale=1)
plt.plot(x_vals, true_pdf, color='red', lw=2, label="True Analytical PDF")

# Step 4: Calculate and plot the CDF
true_cdf = norm.cdf(x_vals, loc=0, scale=1)
plt.plot(x_vals, true_cdf, color='green', lw=2, linestyle='--', label="CDF")

# Step 5: Calculate the sample mean and variance
sample_mean = np.mean(X)
sample_variance = np.var(X)

# Display results
plt.title("Histogram, Analytical PDF, and CDF of Standard Gaussian Random Variable")
plt.xlabel("X")
plt.ylabel("Density / Cumulative Probability")
plt.legend()
plt.grid(True)

# Print sample mean and variance
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Finalize the plot
plt.show()
```

b) Repeat (a) for a Gaussian random variable $Y$ with $mean=3$ and $variance=4$.

```python
# Step 1: Generate a Gaussian random variable Y with mean=3 and variance=4
num_samples = 10000  
mean = 3  
variance = 4  
std_dev = np.sqrt(variance)  # Standard deviation of Y
Y = np.random.normal(loc=mean, scale=std_dev, size=num_samples)  

# Step 2: Plot the histogram (estimated PDF) of Y
plt.figure(figsize=(10, 6))
plt.hist(Y, bins=50, density=True, alpha=0.6,  label='Histogram (Estimated PDF)')

# Step 3: Compute and plot the true analytical PDF of Y
x_values = np.linspace(mean - 5 * std_dev, mean + 5 * std_dev, 1000)  # Range of x values for the true PDF
true_pdf_y = norm.pdf(x_values, loc=3, scale=2) 
plt.plot(x_values, true_pdf, 'r-', lw=2, label='True Analytical PDF')

# Step 4: Calculate the sample mean and variance
sample_mean = np.mean(Y)
sample_variance = np.var(Y)

# Display results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Finalize the plot
plt.title('Histogram and True PDF of Gaussian Random Variable Y (Mean=3, Variance=4)')
plt.xlabel('Y')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

C) Repeat (a) for an exponential random variable with $\lambda = 1$ using [numpy.random.exponential](https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html), and [scipy.stats.expon](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon)

```python


# Step 1: Generate an exponential random variable with =1
lambda_param = 1  # Rate parameter ()
num_samples = 10000  
X = np.random.exponential(scale=1/lambda_param, size=num_samples)  

# Step 2: Plot the histogram (estimated PDF) of the exponential random variable
plt.figure(figsize=(10, 6))
plt.hist(X, bins=50, density=True, alpha=0.6,  label='Histogram (Estimated PDF)')

# Step 3: Compute and plot the true analytical PDF using scipy.stats.expon
x_values = np.linspace(0, 5, 1000)  
true_pdf = expon.pdf(x_values, scale=1/lambda_param) 
plt.plot(x_values, true_pdf, 'r--', lw=2, label='True Analytical PDF')

# Step 4: Calculate the sample mean and variance
sample_mean = np.mean(X)
sample_variance = np.var(X)

# Display results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Finalize the plot
plt.title('Histogram and True PDF of Exponential Random Variable (=1)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 1: Generate an exponential random variable with =1
X = np.random.exponential(scale=1/lambda_param, size=num_samples) 

# Step 2: Plot the histogram (estimated PDF) of the exponential random variable
plt.figure(figsize=(12, 6))

# Plot the histogram for PDF
plt.subplot(1, 2, 1)  # Subplot for PDF
plt.hist(X, bins=50, density=True, alpha=0.6, label='Histogram (Estimated PDF)')

# Step 3: Compute and plot the true analytical PDF using scipy.stats.expon
x_values = np.linspace(0, 5, 1000)  
true_pdf = expon.pdf(x_values, scale=1/lambda_param) 
plt.plot(x_values, true_pdf, 'r-', lw=2, label='True Analytical PDF')

# Step 4: Compute and plot the true CDF using scipy.stats.expon
true_cdf = expon.cdf(x_values, scale=1/lambda_param)  # True CDF of exponential distribution
plt.plot(x_values, true_cdf, 'b-', lw=2, label='True Analytical CDF')

# Step 5: Calculate the sample mean and variance
sample_mean = np.mean(X)
sample_variance = np.var(X)

# Display results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Finalize the plot
plt.title('PDF of Exponential Random Variable (=1)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

plt.show()
```

e) Repeat for an binomial random variable with $n = 5$ and $p = 0.5$

```python
# Step 1: Generate a binomial random variable with n=5, p=0.5
n = 5  
p = 0.5  
num_samples = 10000 
X = np.random.binomial(n=n, p=p, size=num_samples) 

# Step 2: Plot the histogram (estimated PMF) of the binomial random variable
plt.figure(figsize=(10, 6))
plt.hist(X, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6,  label='Histogram (Estimated PMF)')

# Step 3: Compute and plot the true analytical PMF using scipy.stats.binom
x_values = np.arange(0, n+1)  
true_pmf = binom.pmf(x_values, n=n, p=p)  
plt.plot(x_values, true_pmf, 'ro', ms=8, label='True Analytical PMF')
plt.vlines(x_values, 0, true_pmf, colors='r', lw=2)

# Step 4: Calculate the sample mean and variance
sample_mean = np.mean(X)
sample_variance = np.var(X)

# Display results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Finalize the plot
plt.title('PMF of Binomial Random Variable (n=5, p=0.5)')
plt.xlabel('X')
plt.ylabel('Probability Mass')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 1: Generate a binomial random variable with n=5, p=0.5

# X = np.random.binomial(n=n, p=p, size=num_samples)  

# Step 2: Plot the histogram (estimated PMF) of the binomial random variable
plt.figure(figsize=(10, 6))

# Plot the histogram for PMF
plt.hist(X, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, label='Histogram (Estimated PMF)')

# Step 3: Compute and plot the true analytical PMF using scipy.stats.binom
x_values = np.arange(0, n+1)  
true_pmf = binom.pmf(x_values, n=n, p=p)  
plt.plot(x_values, true_pmf, 'ro', ms=8, label='True Analytical PMF')
plt.vlines(x_values, 0, true_pmf, colors='r', lw=2)

# Step 4: Compute and plot the true CDF using scipy.stats.binom (step-like curve)
true_cdf = binom.cdf(x_values, n=n, p=p)  
plt.step(x_values, true_cdf, 'b--', where='post', lw=2, label='True Analytical CDF')

# Step 5: Calculate the sample mean and variance
sample_mean = np.mean(X)
sample_variance = np.var(X)

# Display results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")

# Finalize the plot
plt.title('PMF and CDF of Binomial Random Variable (n=5, p=0.5)')
plt.xlabel('X')
plt.ylabel('Probability / Cumulative Probability')
plt.legend()
plt.grid(True)
plt.show()
```
